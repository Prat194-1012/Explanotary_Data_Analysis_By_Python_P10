```python
# --EXPLORATARY DATA ANALYSIS--
```

```python
#GENDER
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender' , data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True)
sns.barplot(x = 'Gender', y='Amount' , data = sales_gen)
```

```python
# FROM THE ABOVE GRAPHS WE CAN SEE THAT MOST OF THE BUYERS ARE FEMALE AND EVEN THE PURCHASING POWER OF FEMALES ARE HIGH THAN MALES
```

```python
#AGE
```

```python
ax = sns.countplot(x = 'Age Group' , data = df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs age group
sales_gen = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y='Amount' , data = sales_gen)
```

```python
# FROM THE ABOVE GRAPHS WE CAN SEE THAT MOST OF THE BUYERS ARE OF AGE GROUP BETWEEN 26-35 YEARS
```

```python
#STATE
```

```python
df.columns
```

```python
#total orders and state from top 10 states
sales_gen = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_gen, x = 'State', y='Orders')
```

```python
#total amount and state from top 10 states
sales_gen = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_gen, x = 'State', y='Amount')
```

```python
#FROM THE ABOVE GRAPHS WE CAN SEE THAT UNEXPECTEDLY MOST OF THE ORDERS ARE FROM UTTAR PRADESH,KARNATAKA AND MAHARASTRA RESPECTIVELY BUT TOTAL SALES/AMOUNT
#IS FROM UTTAR PRADESH,KARNATAKA AND THEN MAHARASTRA 
```

```python
# MARTIAL STATUS
```

```python
ax = sns.countplot(x = 'Marital_Status' , data = df)

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_gen, x = 'Marital_Status', y='Amount',hue='Gender')
```

```python
#FROM THE ABOVE GRAPH WE CAN SEE THAT MOST OF THE BUYERS ARE MARRIED(WOMEN) AND THEY HAVE HIGH PURCHASING POWER
```

```python
#OCCUPATION
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Occupation' , data = df)

sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_gen, x = 'Occupation', y='Amount')
```

```python
# FROM THE ABOVE GRAPH WE CAN SEE THAT MOST OF THE BUYERS WORKING IN AVIATION AND HEALTHCARE SECTOR
```

```python
# PRODUCT CATEGORY
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Product_Category', data = df)

sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data = sales_gen, x = 'Product_Category', y='Amount')
```

```python
# from the above graph we can see that most of the sold product category are from food,footwear and electronics category
```
